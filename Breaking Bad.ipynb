```python
import wordcloud
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.misc import imread, imfilter
from pylab import rcParams
rcParams['figure.figsize'] = 20, 40
```

```python
text = open('BB_wiki.txt').read()
WW_mask = imread('BB_mask.png')
font_path = 'Heart Breaking Bad.otf'
```

```python
# soften edges of mask and plot
# then make wordcloud with transparent background
plt.imshow(imfilter(WW_mask, 'blur'), alpha=.8)
wc = wordcloud.WordCloud(background_color=None, mode='RGBA', font_path=font_path, mask=WW_mask).generate(text)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
